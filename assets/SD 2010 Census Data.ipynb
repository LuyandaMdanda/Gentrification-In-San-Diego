# Analyzing and Visualizing the San Diego 2010 Census

```python
# Imports - Using these packages to analyze and visualize
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import patsy
import statsmodels.api as sm
import scipy.stats
```

```python
#Reading census data
df = pd.read_excel('SD.xlsx')
```

```python
#Making dataframe with relevant columns and grouping by NeighborhoodName
main_df = df[['NeighborhoodName','TotalPopulation', 'PovertyRt', 'Unemployment', 'Uninsured']]
main_df = main_df.dropna()
main_df.columns.values[0] = 'CPA' 
main_df['Year'] = '2010'

main_df = main_df[['CPA', 'Year', 'TotalPopulation','PovertyRt', 'Unemployment', 'Uninsured']].groupby(['CPA'], as_index=False).mean()
```

```python
fig = pd.scatter_matrix(main_df)
```

```python
x = main_df['TotalPopulation']
y = main_df['PovertyRt']
plt.plot(x, y, "o")
```

```python
x = main_df['Unemployment']
y = main_df['PovertyRt']
plt.plot(x, y, "o")
```

```python
x = main_df['TotalPopulation']
y = main_df['Unemployment']
plt.plot(x, y, "o")
```

```python
x = main_df['Uninsured']
y = main_df['PovertyRt']
plt.plot(x, y, "o")
```

```python
x = main_df['Unemployment']
y = main_df['Uninsured']
plt.plot(x, y, "o")
```

```python
other_df = pd.read_excel("sandag.xlsx", sheetname="Age")
other_df = other_df[['CPA', 'POPULATION']].groupby(['CPA'], as_index=False).mean()

test_df = pd.merge(other_df, main_df, on='CPA')
```

```python
test_df.plot(x='CPA', y='PovertyRt', kind='bar')
```

```python
test_df.plot(x='CPA', y='Unemployment', kind='bar', color = 'green')
```

```python
test_df.plot(x='CPA', y='Uninsured', kind='bar', color='orange')
```
